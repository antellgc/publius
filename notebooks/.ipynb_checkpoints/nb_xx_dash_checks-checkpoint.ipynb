```python
import pandas as pd
```

```python
# only the testing docs can be viewed in app
df_test_docs = pd.read_csv("../processed_data/test_docs.csv")
df_fed_para = pd.read_csv("../processed_data/fed_paragraphs.csv")
df_doc_test_predictions = pd.read_csv("../processed_data/doc_test_predictions.csv")
```

```python
df_test_docs['Indicator Name'] = "Federalist No. " +  df_test_docs['No.'].map(str)
df_test_docs['Full Name'] = df_test_docs['Indicator Name'] + ' - ' + df_test_docs['Title']
df_test_docs_sorted = df_test_docs.sort_values('No.')
df_test_docs_sorted = df_test_docs_sorted.reset_index().drop(['index'],axis=1)
df_test_docs_sorted.to_csv("../processed_data/test_docs_sorted.csv", index=False)

available_indicators = list(df_test_docs_sorted['Indicator Name'])
```

```python
df_merged = pd.merge(df_test_docs_sorted, df_doc_test_predictions)
df_merged.to_csv("../processed_data/test_docs_merged.csv", index=False)
df_merged.head()
```
